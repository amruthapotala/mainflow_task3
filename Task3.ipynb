```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
path="/content/Books.csv"
data = pd.read_csv(path)
```

```python
display(data.head(10))
```

```python
#Scatter plot with Rating against price
plt.scatter(data['Rating'], data['Price'])
#adding title to the plot
plt. title("Scatter Plot")
#setting the xyand y Labels
plt.xlabel('Rating')
plt.ylabel( 'Price')
#adding the Legends
plt.show()
```

```python
#Line chart with Rating against price
plt.plot(data['Rating'])
plt.plot(data['Price'])
#adding title to the plot
plt. title("Line Chart")
#setting the xyand y Labels
plt.xlabel('Rating')
plt.ylabel( 'Price')
#adding the Legends
plt.show()
```

```python
#Bar chart with RAM against Final price
plt.bar(data['Rating'], data['Price'])
#adding title to the plot
plt. title("Bar Chart")
#setting the xyand y Labels
plt.xlabel('Rating')
plt.ylabel( 'Price')
#adding the Legends
plt.show()
```

```python
#Histogram
plt.hist(data['Rating'])
plt.title("Histogram")
plt.show( )
```
